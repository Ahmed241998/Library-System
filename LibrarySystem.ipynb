```python
# Creation of a class
class Library :
    # dictionary will have the name of the book as key and the user who borrowed  the book as value
    dct= {}
    # Intitiation function of the class 
    def __init__(self,book_list,library_name) :
    # Class atribute which contains the list of books
        self.book_list=book_list
    # Class attribute which contain the name of the library
        self.library_name=library_name
    # Method which will return all available books for lending
    def available_books (self):
        print(self.book_list)
    # Method will help the user to add books to his book list and update it
    def adding_books (self,add_books) :
        self.book_list.extend(add_books)
    # Method will help the user to remove the lended book from the available book list and upadte the book:user dictionary
    def lending_book(self,lend_book,user) :
        self.book_list.remove(lend_book)
        Library.dct [lend_book]=user
    # Method will add the returned book to available book for lending once again and delete this book from book:user dictionary
    def returning_book(self,returned_book) :
        self.book_list.append(returned_book)
        del Library.dct[returned_book]
```

```python
def userchoice () :
    """
    Ask user to select between 1, 2, 3, 4
    Returns :
    (float) user_choice : the user choice between 1 ,2, 3, 4 
    """
    # Use a while loop to handle invalid inputs and ask user for correct selection
    print("1.Display available books.\n2.lending books.\n3.Add book to library.\n4.Return book.")
    while True :
        try :
            user_choice= float(input("Please Enter Your choice as a number 1,2,3 or 4 : "))    
            if user_choice in [1,2,3,4] :
                break
        except :
            print("Enter Correct choice")
    return user_choice
```
